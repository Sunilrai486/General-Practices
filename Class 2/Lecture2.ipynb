```python
from sklearn.datasets import make_classification
features, target = make_classification(n_samples = 100, 
                                                n_features = 3, 
                                                n_informative = 2, 
                                                n_redundant = 0,
                                                n_classes = 2,
                                                weights = [.25, .75],
                                                random_state = 4)
```

```python
print('Feature Matrix\n', features[:3])
print('Target Vector \n', target[:3])
```

```python
features.dtype
```

```python
target.dtype
```

```python
import numpy as np
features = np.array(features)
target = np.array(target)
mypd = pd.DataFrame(data=[features[0], features[1], features[2], target]).T
mypd.columns=['features', 'features2', 'features3',  'target']
print(mypd)
```
